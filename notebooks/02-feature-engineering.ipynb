```python
import pandas as pd
import numpy as np
```

```python
#Load cleaned data from previous step
spotify_df = pd.read_csv('../data/processed/spotify_user_behavior_cleaned.csv')
test = pd.read_csv('../data/processed/spotify_user_behavior_cleaned.csv')

print(f"Loaded data shape: {spotify_df.shape}")
print(f"\nColumns: {list(spotify_df.columns)}")
```

### User Preference Features

```python
#Create new binary column from most_liked_feature
spotify_df['likes_personalization'] = spotify_df['most_liked_feature'].apply(lambda x: 1 if x in ['Daily Mix', 'Discover Weekly', 'AI DJ'] else 0)
print(spotify_df['likes_personalization'].value_counts())

#Create new feature for users who dislike music suggestions
spotify_df['dislikes_suggestions'] = (spotify_df['music_suggestion_rating_1_to_5'] <= 2).astype(int)
print(spotify_df['dislikes_suggestions'].value_counts())
```

### Usage Features

```python
#Create new feature for heavy listeners
spotify_df['heavy_listener'] = (spotify_df['avg_listening_hours_per_week'] > 15).astype(int)
print(spotify_df['heavy_listener'].value_counts())

#Create new feature based on number of playlists created
spotify_df['many_playlists'] = (spotify_df['playlists_created'] > 10).astype(int)
print(spotify_df['many_playlists'].value_counts())

#Create new feature to identify new/old users from signup date(within last 2 years)
spotify_df['new_user'] = (spotify_df['signup_date'] >= '2024-01-01').astype(int)
print(spotify_df['new_user'].value_counts())
```

### Save data with new features

```python
spotify_df.to_csv('../data/processed/spotify_user_behavior_features.csv', index=False)
```
